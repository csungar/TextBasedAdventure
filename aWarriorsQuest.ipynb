```python
import random


# introduction to our game.
def intro():
    """
         Welcome to my text based adventure: A warriors quest.
         This game is based on Dungeons & Dragons or any other tabletop RPG game and
         MMORPG World of Warcraft.

         It will have 4 stages:
         1)Intro
         2)tavern
         3)taking wizards quest
         4)cavern
     """
    # our starting zone is randomized with random.choice here
    start_zone = random.choice(['Azeroth', 'Outland', 'Dreanor', 'Kalimdor'
                                'Pandaria', 'Broken Isles'])
    print(f"""
         The sun rises upon the Mystic Land of {start_zone}. You are a mighty
         warrior looking for an adventure. There are rumors going
         around about a wizard arriving to your town: Goldshire. You will
         find this wizard in your town's tavern: The World's End Tavern.
         """)
    # Our message when game is complete or when game is lost 'Quitting' is here
    prompt("Press enter to continue..")
    tavern()
    print("Quiting..")


# meeting the wizard in the taverns
def tavern():
    """
     We meet the wizard in the tavern.
    """
    # dynamic menu for the first user selection is defined
    first_menu_str  = "1) Sit at the table that the wizard is sitting at."
    second_menu_str = "2) Stand in a corner alone."
    third_menu_str  = "3) Leave."

    # a random hour between 8 and 11 is assingned as our hour
    selected_hour = random.randint(8,11)
    # choose what to do in the tavern
    print(f"""
          You enter The World Ends tavern at {selected_hour}.00 am. The place seems quite crowded.
          You spot the wizard sitting in the tavern.
          What would you like to do next?

          {first_menu_str}
          {second_menu_str}
          {third_menu_str}

    """)

    # asking user to make a choice.
    while True:
        choice = input('prompt = ')
# every time check_user_input is called it lower cases all input, and checks whether input is suitable
        try:
            if check_user_input(choice, first_menu_str, first_menu_str.split(" ") + ["1", "one", "first"]):
                prompt("You sit next to the wizard. Press enter to continue..")
                wizards_quest()
                break

            elif check_user_input(choice, second_menu_str, second_menu_str.split(" ") + ["2", "two", "second"]):
                prompt("You proceed to a corner. Press enter to continue..")
                meet_allies()
                break

            elif check_user_input(choice, third_menu_str, third_menu_str.split(" ") + ["3", "three", "third"]):
                game_over("You have made your own destiny :/ ")
                break

            else:
                exception_error(f"{choice} is an unrecognized input!")
        # exception goes here
        except Exception as ex:
            print(f"Exception occurred. Details: {ex}")


# our quest is being taken from the wizard
def wizards_quest():
    first_menu_str  = "1) Raise your hand and accept the quest."
    second_menu_str = "2) Remain quiet."
    third_menu_str  = "3) Attack the wizard."
    # Lord Cobrahn is assigned a random class here
    selected_class = random.choice(['druid', 'mage', 'warlock', 'rogue', 'shaman'])
    # wizard announces his quest here
    # dynamic menu is printed after wizard presents his quest
    print(f"""
          Wizard starts talking out loud and proceeds to announce his quest:
          Adventurers! I have dangerous quest at hand for those who dare. You
          have to travel to Wailing Caverns to defeat a mighty {selected_class} called
          Lord Cobrahn. Who is powerful enough to take on this quest?

          {first_menu_str}
          {second_menu_str}
          {third_menu_str}

    """)
# in check_user input Included arguments are presented inside brackets
    while True:
        answer = input('prompt = ')
        try:
            if check_user_input(answer, first_menu_str, first_menu_str.split(" ") + ["1", "one", "first"]):
                prompt("You are off to the cave with your allies. Press enter to continue..")
                wailing_caverns()
                break

            elif check_user_input(answer, second_menu_str, second_menu_str.split(" ") + ["2", "two", "second"]):
                game_over("You have missed the adventure :( ")
                break

            elif check_user_input(answer, third_menu_str, third_menu_str.split(" ") + ["3", "three", "third"]):
                game_over("The wizard freezes you in place.")
                break

            else:
                exception_error(f"'{answer}' input is unrecognized!")

        except Exception as ex:
            print(f"Exception occurred. Details: {ex}")


# if we meet our allies here is where we meet them
def meet_allies():
    first_menu_str  = "1) Accept."
    second_menu_str = "2) Decline."

    # a random shuffle of allies will return to our function
    ally_list = ['an orc', 'an elf', 'a gnome', 'a dwarf', 'a blood elf', 'a tauren', 'an undead']
    random.shuffle(ally_list)
    # shuffle created a new list, now we select 3 elements randomly
    ally_1 = ally_list[0]
    ally_2 = ally_list[1]
    ally_3 = ally_list[2]
    # our selection is printed dynamically in our print
    print(f"""
          A trio of {ally_1}, {ally_2} and {ally_3} sit next to you and invite you
          to join their party.

          If you accept you will start obeying their leader..

          {first_menu_str}
          {second_menu_str}
          """)

    while True:
        ally_answer = input('prompt = ')
        try:
            if check_user_input(ally_answer, first_menu_str, first_menu_str.split(" ") + ["1", "one", "first"]):
                print("""Their band decided to take on the adventure. You're off to the cave.""")
                wailing_caverns()
                break
            # if we decline their offer we die
            elif check_user_input(ally_answer, second_menu_str, second_menu_str.split(" ") + ["2", "two", "second"]):
                game_over("They took offense and cut your arm off. :{ ")
                break
            # returns exception
            else:
                exception_error(f"{ally_answer} is an unknown!")

        except Exception as ex:
            print(f"Exception occurred. Details: {ex}")

# last stage of our game
def wailing_caverns():
    first_menu_str  = "1) Cast a dispel on your ally."
    second_menu_str = "2) Attack Cobrahn."
    third_menu_str  = "3) Run away."

    # list for different possible allies
    allies = ["a mage and a warlock",
              "a rogue and a druid",
              "a shaman and a monk",
              "a warrior and a death-knight"]
    # random ally duo returns from list above [1,4)
    random_allies = allies[random.randrange(1, 4)]
    print(f"You enter wailing caverns with your allies: {random_allies}.\n"
          f"You proceed into the depths of the cave and encounter Lord Cobrahn.")
    
    # you have to prioritize dispeling your ally here else you lose the game
    print(f"""
          Lord Cobrahn gets the first move and casts a polymorph on your
          ally.
          Pick an action..

          {first_menu_str}
          {second_menu_str}
          {third_menu_str}
         
    """)
# 'prompt' is used everytime we take input
    while True:
        last_answer = input('prompt =')
        try:
            if check_user_input(last_answer, first_menu_str, first_menu_str.split(" ") + ["1", "one", "first"]):
                print("Your ally is free to attack again.\nIt is your turn to attack now")
                attack_cobrahn()
                break

            elif check_user_input(last_answer, second_menu_str, second_menu_str.split(" ") + ["2", "two", "second"]):
                game_over("Cobrahn killed your ally. And parried your attack")
                break

            elif check_user_input(last_answer, third_menu_str, third_menu_str.split(" ") + ["3", "three", "third"]):
                game_over("Cobrahn's minions blocked your way. Your whole party is disabled.")
                break

            else:
                exception_error(f"{answer} is an unknown!")
# if an exception arises it goes here
        except Exception as ex:
            print(f"Exception occurred. Details: {ex}")


# we fight our final boss here
def attack_cobrahn():
    # random number of lives between 1 and 10 returns as lives count of the boss
    lives = round(random.uniform(1,10))
    print(f"Lord Cobrahn has {lives} lives.\nHit 'x' to launch an attack!")
# number of lives we have will be reduced by 1 each time we press 'x' key in this while loop
    while lives != 0:
        attack = input('prompt = ')
        try:
            if not check_user_input(attack, 'x', ['x']):
                raise Exception("Please press 'x' to attack!")

            lives -= 1
            if lives == 0:
                print("You have defeated Cobrahn. Quest complete!")
            else:
                print("Cobrahn has ", lives, "lives left.")
        except Exception as inst:
            print(f"Invalid input. Details: {inst}")


# a prompt that is dynamic with user input
def prompt(message):
    print(message)
    input('prompt = ')


# a game over prompt with dynamic user input
def game_over(message):
    """
    :param : (str) message Game over status
    """

    prompt(f"Game over! {message}. Press enter key to exit!")


# exception handling generic
def exception_error(message):
    raise Exception(message)


# normalize string into lowercase
def normalize_string(input_string):
    return input_string.lower()


# Validate User input for the given arguments
def check_user_input(user_input, check_expression, includes=[], excludes=[" "]):
    """
    :param user_input:(str) User input
    :param check_expression:(str) We split this expression using whitespaces into an array and search for user_input in it.
    :param includes:(list) Additional expressions list to check user_input in them,
    :param excludes:(list) Excluded words list from the checks
    :return:(bool) Indicating whether the given user_input is contained.
    """
    if user_input == "":
        raise Exception("User Input Can Not be Empty!")

    # before we perform the checks lets lowercase everything
    user_input = normalize_string(user_input)
    check_expression = normalize_string(check_expression)

    # extend the check_expression with the included words
    check_expression_extended = check_expression + normalize_string(normalize_string(" ".join(includes)))

    for exclude in excludes:
        exclude = normalize_string(exclude)
        check_expression_extended = check_expression_extended.replace(exclude, "")

    if user_input in check_expression_extended:
        return True
    else:
        # now, lets additionally check if there is any included words which the user_input might contain
        for exp in includes:
            if exp in user_input:
                return True
        return False


# program start
intro()


# References
#https://stackoverflow.com/questions/12309976/how-do-i-convert-a-list-into-a-string-with-spaces-in-python/12309982
#https://www.tutorialspoint.com/python/string_replace.htm
#https://www.w3schools.com/python/gloss_python_join_lists.asp
#https://www.pythonforbeginners.com/random/how-to-use-the-random-module-in-python
#https://docs.python.org/3/library/random.html
#https://www.geeksforgeeks.org/python-random-module/
#https://www.tutorialsteacher.com/python/random-module
#https://stackabuse.com/how-to-randomly-select-elements-from-a-list-in-python/
```
